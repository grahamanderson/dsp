```python
import numpy as np
import pandas as pd
df = pd.read_json('https://api.github.com/repos/mbostock/d3/stats/commit_activity')
df.head()
```

# Question 1: What Week had the highest number of commits?
# Answer: Week of Sunday October 18

```python
import datetime
highest_row = df.ix[df['total'].idxmax()]
week= highest_row["week"] #timestamp convert  Sun, 18 Oct 2015 00:00:00 GMT
print(week.dtype)
print(highest_row)
print(week)
#print(test)
```

# Question: What Day had the Most Commits?
# Answer: Wedneday  (from chart below)

```python
# Note: might have been better to isolate the json node daysover reading the  entire json string in...
df.days.apply(lambda s: pd.Series({'sun':s[0], 'mon':s[1],'tues':s[2], 'wed':s[3], 'thurs':s[4], 'fri':s[5], 'sat':s[6] }))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
days = df.days.apply(lambda s: pd.Series({'sun':s[0], 'mon':s[1],'tues':s[2], 'wed':s[3], 'thurs':s[4], 'fri':s[5], 'sat':s[6] }))
# in som future-esque iteration, order the columns alphabetically
print(days.sum(axis=0))
```

```python
totals = days.sum(axis=0)
totals.plot(kind="bar")
```
